```python
from sqlalchemy import create_engine
import os
import numpy as np
import pandas as pd
from dotenv import load_dotenv

def create_connection():

    load_dotenv()
    host = os.environ.get('DB_DESTINATION_HOST')
    port = os.environ.get('DB_DESTINATION_PORT')
    db = os.environ.get('DB_DESTINATION_NAME')
    username = os.environ.get('DB_DESTINATION_USER')
    password = os.environ.get('DB_DESTINATION_PASSWORD')
    
    conn = create_engine(f'postgresql://{username}:{password}@{host}:{port}/{db}')
    return conn

# установите соединение с базой
conn = create_connection()

data = pd.read_sql('select * from flat_target_price', conn)
```

```python
data=pd.read_sql(f"SELECT * FROM flat_target_price", conn)
print(data)
```

```python
is_duplicated_id = data.duplicated(subset=['id'], keep=False)
print(is_duplicated_id)
print(sum(is_duplicated_id)) 
```

```python
feature_cols = [col for col in data.columns if col != 'id']
is_duplicated_features = data.duplicated(subset=feature_cols, keep=False)
print(sum(is_duplicated_features))
print(data[is_duplicated_features].sort_values(by=feature_cols))
```

```python
data = data[~is_duplicated_features].reset_index(drop=True) 
print(data)
```
